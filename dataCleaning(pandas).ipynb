```python
import pandas as pd

df = pd.read_excel("Customer Call List.xlsx")
df
```

```python
df=df.drop_duplicates().copy() #removing duplicate rows and creating a copy
df
```

```python
#deleting unneccessary data
df.drop(columns="Not_Useful_Column",inplace=True)
df
```

```python
# cleaning last_name column
#strip - first method
# df['Last_Name'] = df['Last_Name'].str.lstrip("...")
# df['Last_Name'] = df['Last_Name'].str.lstrip("/")
# df['Last_Name'] = df['Last_Name'].str.rstrip("_")
 #Strip() strips from the both left and right side where as [lstrip()-removes left side part],[rstrip()- removes right side part]
# df

#instead of following the above syntax, you can specify all the items you need to delete alltogether
df['Last_Name']=df['Last_Name'].str.strip("123._/")
df
```

```python
#to format the phone numbers into a consistent format
# Remove all non-digit characters
df.loc[:, 'Phone_Number'] = df['Phone_Number'].astype(str).str.replace(r'\D', '', regex=True)

# Optional: Further format 10-digit numbers (e.g., as 'XXX-XXX-XXXX')
# This part is optional and depends on your desired final format
# df.loc[df['Phone_Number'].str.len() == 10, 'Phone_Number'] = df['Phone_Number'].str.replace(r'(\d{3})(\d{3})(\d{4})', r'\1-\2-\3', regex=True)
df
# str(df['Phone_Number'])
```

```python
#now to format the cleaned numbers, we can use lambda function or for loop
df["Phone_Number"]=df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6]+ '-'+x[6:10])
df
```

```python
df[["Street_Address", "State", "Zip_Code"]]=df["Address"].str.split(',', n=2, expand=True)
df
```

```python
df['Paying Customer'] = df['Paying Customer'].replace({'Yes': 'Y', 'No': 'N'}).fillna('N')
df['Paying Customer'] = df['Paying Customer'].str.strip() # Remove any leading/trailing whitespace
df
```

```python
# df=df.replace('N/a','')
# df=df.replace('NaN','')   #OR

df=df.fillna('')
df
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].replace({'Yes': 'Y', 'No': 'N'}).fillna('N')
df
```

```python
for i in df.index:
  if df.loc[i,"Do_Not_Contact"]=='Y':
    df.drop(i,inplace=True)
df
```

```python
for i in df.index:
  if df.loc[i,"Phone_Number"]=='--':
    df.drop(i,inplace=True)
df
```

```python
df.reset_index(drop=True,inplace=True)
df
```
